```python
# Calcular estadísticas descriptivas para cada género
classical_stats = classical_emotions[columns_emotions].describe()
rock_stats = rock_emotions[columns_emotions].describe()
electronic_stats = electronic_emotions[columns_emotions].describe()
pop_stats = pop_emotions[columns_emotions].describe()

# Combinar las estadísticas en un solo dataframe
stats_combined = pd.concat({
    'classical': classical_stats,
    'rock': rock_stats,
    'electronic': electronic_stats,
    'pop': pop_stats
}, axis=1)

# Mostrar el dataframe combinado
stats_combined
```

```python
# Seleccionar las estadísticas principales que queremos visualizar
stats_variables = ["mean", "std", "min", "max"]

# Crear una figura con subplots
fig, axs = plt.subplots(2, 2, figsize=(18, 12))

# Graficar cada estadística en un subplot diferente
for stat, ax in zip(stats_variables, axs.flat):
    graphs = stats_combined.loc[stat]
    
    # Crear un gráfico de barras agrupado
    graphs.plot(kind="bar", ax=ax, colormap="viridis")
    ax.set_title(f"Comparación de {stat}")
    ax.set_xlabel("Emociones")
    ax.set_ylabel(f"{stat.capitalize()}")
    ax.tick_params(axis="x", rotation=45)
    ax.grid(axis="y", linestyle="--", alpha=0.5)

plt.tight_layout()

```

##### Análisis de los principales estadísticos de la Edad:

Count: 8407 Este número indica la cantidad de registros. Es decir hay 8407 personas cuyas edades están registradas.

Mean (media): 31.85 La media es el promedio de las edades. Esto nos dice que la edad promedio de las personas en este conjunto de datos es de aproximadamente 31.85 años.

Std (desviación estándar): Mide la dispersión o variabilidad de las edades respecto a la media. Una desviación estándar de 12.46 sugiere que las edades de las personas varían bastante alrededor de la media.

Min (mínimo): 5 es la edad mínima.

Primer cuartil (25%): El primer cuartil representa el valor por debajo del cual se encuentra el 25% de los datos. En este caso, el 25% de las personas tienen 23 años o menos.

Mediana (50%): La mediana es el valor central que divide los datos en dos mitades iguales. Aquí, la mediana es de 28 años, lo que significa que la mitad de las personas tienen 28 años o menos y la otra mitad tiene más de 28 años.

Tercer cuartil (75%): El tercer cuartil representa el valor por debajo del cual se encuentra el 75% de los datos. En este caso, el 75% de las personas tienen 37 años o menos.

Max (máximo): La edad máxima registrada en el conjunto de datos es de 99 años.

Explicación del Diagrama de Cajas por estilo musical y edades:

Eje X (Estilo Musical): Se encuentran los diferentes estilos musicales (clásico, rock, electrónico, pop).

Eje Y (Edad): Edades de todas las personas.

Forma del Diagrama de Cajas:

* Cajas: muestra la dispersión de edades entre el el primer cuartil (Q1 25%) y el tercer cuartil (Q3 75%).

* Líneas o bigotes: muestran el rango de edades, lo que esta fuera serían los valores atípicos.

* Valores atípicos (Outsiders): valores fuera de los bigotes/líneas que son más altos o bajos que el resto.

* Mediana: es la línea horizontal que representa el 50% (Q2), la mitad de los valores se encuentran por encima o por debajo de esa línea. El valor central.
